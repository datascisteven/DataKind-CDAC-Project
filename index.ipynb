## Ookla Database

The column keys are:
- `avg_d_kbps`, `int`, average download speed of all tests performed in the tile, represented in kilobits per second.
- `avg_u_kbps`, `int`, 	average upload speed of all tests performed in the tile, represented in kilobits per second.
- `avg_lat_ms`, `int`, average latency of all tests performed in the tile, represented in milliseconds
- `tests`, `int`, number of tests taken in the tile.
- `devices`, `int`, number of unique devices contributing tests in the tile.
- `quadkey`: quadkey representing the tile.



````python
# import Ookla database
ny_3 = pd.read_csv('/content/drive/MyDrive/Datasets/Broadband/NY/ookla_combined_ny.csv')
ny_3.head()
````

````python
il_shp = gpd.read_file("/content/drive/MyDrive/Datasets/Broadband/IL/il_spdf/il_spdf.shp")
il_shp = il_shp.sort_values(by=['GEOID'])
il_shp.head()
````

````python
il_1 = pd.read_csv('/content/drive/MyDrive/Datasets/Broadband/IL/acs_2019_IL.csv')
il_1 = il_1.rename(columns={'geoid':'GEOID'})
il_1['GEOID'] = il_1['GEOID'].apply(lambda x: str(x))
il_1.head()
````

````python
il_2 = pd.read_csv('/content/drive/MyDrive/Datasets/Broadband/IL/fcc_477_census_tract_IL.csv')
il_2 = il_2.rename(columns={'tract':'GEOID'})
il_2['GEOID'] = il_2['GEOID'].apply(lambda x: str(x))
il_2.head()
````

````python
il_3 = pd.read_csv('/content/drive/MyDrive/Datasets/Broadband/IL/ookla_combined_il.csv')
il_3a = il_3.GEOID.value_counts().rename_axis('GEOID').reset_index(name='counts')
il_3a = il_3a.sort_values(by=['GEOID'])
il_3 = il_3.groupby('GEOID', as_index=False).agg({'tests': 'sum', 'devices': 'sum', 'avg_d_kbps': 'mean', 'avg_u_kbps': 'mean', 'avg_lat_ms': 'mean'})
il_3 = il_3.merge(il_3a, on="GEOID")
il_3['GEOID'] = il_3['GEOID'].apply(lambda x: str(x))
il_3.head()
````
